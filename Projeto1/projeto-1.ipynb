```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# <center>  Exercício 1 </center>

## Dados de casos de covid

```python
df_casos_covid = pd.read_csv('data/casesBrazil.csv')
```

```python
df_casos_covid
```

```python
fig, ax = plt.subplots(1,1)

ax.scatter(df_casos_covid['day'], df_casos_covid['cases'], color='orange')
ax.set_xlabel('number of days', fontsize=10)
ax.set_ylabel('amount of cases', fontsize=10)

plt.show()
```
